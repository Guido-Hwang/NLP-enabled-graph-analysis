### 将csv文件中符合条件的的用户抽取出来
### 防止重复爬取用户

```python
import pandas as pd
from tqdm import tqdm
import os
import re
floor=11
key = 'author_user_name'
user= pd.read_csv('user_L{0}_post_comment.csv'.format(floor),encoding='utf_8_sig',usecols=[key])
user.dropna(inplace=True)
```

### 过滤已经爬取的用户
### 求差集

```python
file_names = os.listdir(os.getcwd())  # 路径
user_file = (re.findall('user_L[0-9].csv', ''.join(file_names)))
print(user_file)
# 将用户数据合成
exist_user_list=[]
for file in user_file:
    exist_user_list.extend( pd.read_csv(file, encoding='utf_8_sig', usecols=[key])[key].tolist())
exist_user_list=set(exist_user_list)
floor_user_list=set(user[key].tolist())
unclawer_user_list=list(floor_user_list-exist_user_list)
print(unclawer_user_list)
print(f'filter {user.shape[0]- len(unclawer_user_list)}')
```

```python
next_floor_user=pd.DataFrame({key:unclawer_user_list})
next_floor_user.to_csv('user_L{0}.csv'.format(floor+1),index=False,encoding='utf_8_sig')
```
